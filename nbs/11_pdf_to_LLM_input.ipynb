```python
from docling.document_converter import DocumentConverter

file_path = "/home/vule/projects/learn_agent/nexus_reflection_data/Key Dimension Guidelines.docx.pdf"
converter = DocumentConverter()
result = converter.convert(file_path)
```

```python
import pytesseract
from PIL import Image
from pdf2image import convert_from_path

# Convert the PDF to images (one image per page)
images = convert_from_path(file_path)

# Extract text from each image using OCR
extracted_text = ""
for img in images:
    extracted_text += pytesseract.image_to_string(img, lang='eng')

# Display the first 2000 characters of the extracted text to confirm output quality
extracted_text[:2000]
```
